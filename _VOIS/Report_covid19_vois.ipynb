```python
#importing pandas
import pandas as pd
import numpy as np
```

```python
#dataframe 'df' out of covid_19_india.csv dataset
df = pd.read_csv('covid_19_india.csv')
df
```

```python
#Dropping the irrelevant columns
df.drop('Sno', axis=1, inplace=True)
df
```

```python
##Dropping the irrelevant columns
df.dropna(axis=1,thresh=1,inplace=True)
df
```

```python
#active patients and add a column 'Active' to the dataframe
df['Active'] = df['Confirmed'] - (df['Cured'] + df['Deaths'])
df
```

```python
max_state_value = df.groupby('State/UnionTerritory')['Active'].sum()
```

```python
#state having current maximum Active patient and noof active patients
max_state_value[max_state_value == max(max_state_value)]
```

```python
#state/union territory having current minimum Active patient and noof active patients
min_state_value = df.groupby('State/UnionTerritory')['Active'].sum()
```

```python
min_state_value[min_state_value == min(min_state_value)]
```

```python
death_to_cure = df.groupby('State/UnionTerritory').sum()
death_to_cure
```

```python
#Deaths to Cured ratio then find the states having
minimum and maximum ratios
death_to_cure['death/cure'] = death_to_cure['Deaths']/death_to_cure['Cured']
death_to_cure
```

```python
#Deaths to Cured ratio then find the states having minimum and maximum ratios
death_to_cure[death_to_cure['death/cure'] == max(death_to_cure['death/cure'])]
```

```python
#Pie chart

#Quantities for plotting
condition = ["Deaths","Active","Cured"]
person = [146572.0,1398338.0,3349493.0]

#Defining colors for plot
colors = ["red","orange","blue"]

# Plotting pie chart
plt.pie(person, labels=condition,explode = [0.1,0,0] ,colors=colors,autopct='%1.1f%%', shadow=True, startangle=120)

#Defining the title of our plot
plt.title("Delhi")

#Show what have been plotted
plt.show()
```
